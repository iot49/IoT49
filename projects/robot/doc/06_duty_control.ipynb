# Duty Controller

Just sets duty cycle, no feedback control.

```python
import nest_asyncio
nest_asyncio.apply()

import asyncio, sys, os
sys.path.append(os.path.join(os.getenv('IOT49'), 'projects/robot/code-pi'))

from robot import RobotComm
from robot import LivePlot
import stm32
import numpy as np

from bokeh.io import output_notebook

output_notebook()

def exception_handler(loop, context):
    msg = context.get("exception", context["message"])
    print("***** asyncio:", context)
    print("***** msg:", msg)
    
async def main(fs=10, N=100):
    async with RobotComm(fs=fs) as rc:
        asyncio.get_event_loop().set_exception_handler(exception_handler)
        # lp = LivePlot('time [s]', ['dt1 [us]', 'dt2 [us]', 'duty1', 'duty2'], rollover=200, title="Duty Cycle")
        lp = LivePlot('duty1', ['diff_enc1', 'diff_enc2'], rollover=200, title="Duty Cycle")
        lp.show(rc.get_state_queue())
        try:
            rc.start_controller(fs, 'duty_control')
            for duty in np.linspace(0, 100, N):
                rc.set_fstate('duty1', duty)
                rc.set_fstate('duty2', duty)
                await asyncio.sleep(1/fs)
            await rc.ping()
            rc.stop_controller()
        finally:
            pass

print("reset")
stm32.hard_reset()
if False:
    stm32.rsync(dry_run=False)
    stm32.hard_reset()

asyncio.run(main(fs=10, N=100))
```

---
**Optional: Unix port**

---

```python
%%bash

cd ~/micropython/ports/unix
make submodules
make
```

```python
%%bash

cd ~/micropython/ports/unix
./micropython -c "import sys; print(sys.platform)"
```

```python
%%bash

cd ~/micropython/ports/unix
./micropython -m upip install micropython-pystone
./micropython -m pystone
```

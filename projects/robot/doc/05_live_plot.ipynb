# LivePlot

```python
import nest_asyncio
nest_asyncio.apply()

import asyncio, sys, os
sys.path.append(os.path.join(os.getenv('IOT49'), 'projects/robot/code-pi'))

from robot import LivePlot

from bokeh.io import output_notebook
import asyncio, queue

output_notebook()

async def data_producer(data_queue):
    """Send data to LivePlot via data_queue"""
    import math
    from random import random
    t = 0
    while True:
        new_data = { 'time': t, 
             'sin':  math.sin(0.1*t),
             'cos':  math.cos(0.1*t) + (math.exp(1e-3*t)-1)*(random()-0.5),
             'extra': 5
           }
        t += 1
        data_queue.put(new_data)
        await asyncio.sleep(0.008)      

lp = LivePlot('time', ['sin', 'cos'], rollover=200, update_interval_ms=10, y_range=(-2, 2), title="Bokeh LivePlot")
data_source = queue.Queue()
lp.show(data_source)

try:
    asyncio.run(data_producer(data_source))
except KeyboardInterrupt:
    print("KeyboardInterrupt")
```

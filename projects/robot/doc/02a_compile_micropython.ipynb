# Compile Micropython

```python
%connect 'serial:///dev/ttyAMA2'
```

#### Download

```python
%%bash

cd ~
git clone https://github.com/iot49/micropython.git
```

#### mpy-cross

```python
%%bash

cd ~/micropython/mpy-cross
make
```

#### UNIX port (optional)

```python
%%bash

apt-get update
apt-get -y install libffi-dev
apt-get -y install pkg-config

cd ~/micropython/ports/unix
make submodules
make
```

```python
%%bash

cd ~/micropython/ports/unix
./micropython -c "print(list(5 * x + y for x in range(10) for y in [4, 2, 1]))"
./micropython -m upip install micropython-pystone
./micropython -m pystone
```

#### STM32 - Motor Hat Binary

```python
%%bash

cd ~/micropython/ports/stm32
make submodules
make BOARD=MOTOR_HAT
```
